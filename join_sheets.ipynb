```python
import pandas as pd
```

```python
sheets = [pd.read_csv("rnn-2022-03-16 13-55-17.csv"), pd.read_csv("dqn-ani2-1hl-2022-03-15 20-32-06.csv"), 
         pd.read_csv("dqn-ani-1hl-2022-03-15 11-13-53.csv"), pd.read_csv("dqn-ani-1hl-2022-03-14 10-49-57.csv"),
         pd.read_csv("dqn-2hl-2022-03-15 14-48-33.csv"), pd.read_csv("dqn-2hl-2022-03-13 16-36-15.csv"),
         pd.read_csv("dqn-1hl-2022-03-15 15-24-53.csv"), pd.read_csv("dqn-1hl-2022-03-14 09-51-58.csv"),
         pd.read_csv("dqn-1hl-2022-03-14 04-42-41.csv"), pd.read_csv("dqn-1hl-2022-03-14 04-07-20.csv"),
         pd.read_csv("dqn-1hl-2022-03-13 22-01-26.csv"), pd.read_csv("dqn-1hl-2022-03-13 20-48-57.csv"),
         pd.read_csv("dqn-1hl-2022-03-13 14-11-15.csv"), pd.read_csv("dqn-1hl-2022-03-12 22-11-56.csv"),
         pd.read_csv("dqn-1hl-2022-03-12 21-34-02.csv"), pd.read_csv("dqn-1hl-2022-03-11 03-55-36.csv"),
         pd.read_csv("dqn-1hl-2022-03-09 04-34-45.csv")]
```

```python
sheets[1].head()
```

```python
def clean_up(df: pd.DataFrame) -> pd.DataFrame:
    
    # remove unnamed initial column.
    # replace model column with name of model
    df = df.drop(["Unnamed: 0"], axis=1)
    if "QNetwork" in df["model"][0]:
        df["ModelType"] = "DQN"
    elif "RNN" in df["model"][0]:
        df["ModelType"] = "RNN"
    df = df.drop(["model"], axis=1)
    
    return df 
```

```python
for idx, sheet in enumerate(sheets):
    print(f"Cleaning up sheet {idx}")
    sheets[idx] = clean_up(sheet)   
```

```python
# delete redundant columns

for sheet in sheets:
    if "Model Type" in sheet.columns:
        sheet.drop("Model Type", inplace=True, axis=1)

for sheet in sheets:
    if "Unnamed: 0.1" in sheet.columns:
        sheet.drop("Unnamed: 0.1", inplace=True, axis=1)
```

```python
columns = []
for idx, sheet in enumerate(sheets):
    for col in sheet.columns:
        if not col in columns:
            print(f"Adding {col} to columns.")
            columns.append(col)
```

```python
# now we join the spreadsheets together

# first we need to make sure all sheets have the same columns
for sheet in sheets:
    for col in columns:
        if not col in sheet.columns:
            sheet[col] = None
```

```python
master_sheet = pd.concat(sheets, ignore_index=True, sort=True)
```
