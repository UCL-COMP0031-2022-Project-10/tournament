```python
def clean_up_tabular(df: pd.DataFrame) -> pd.DataFrame:
    df = df.drop(["Unnamed: 0", "model"], axis=1)
    
    return df 

def map_model(x: str) -> str:
    if "GRU" in x:
        return "GRU"
    
    return "RNN" if "RNN" in x else "DQN"

def clean_up_qnetwork(df: pd.DataFrame) -> pd.DataFrame:
    df = df.drop(["Unnamed: 0"], axis=1)
    df["model"] = df["model"].apply(map_model)
    df = df.rename(columns={"model" : "ModelType"})
    
    return df
```

```python
print("Cleaning Q-Network Sheets")
for idx, qnetwork_sheet in enumerate(qnetwork_sheets):
    qnetwork_sheets[idx] = clean_up_qnetwork(qnetwork_sheet)   
print("Finished cleaning Q-Network")

print("Cleaning tabular")
for idx, tab_sheet in enumerate(tab_sheets):
    tab_sheets[idx] = clean_up_tabular(tab_sheet)   
print("Finished cleaning tabular")
```

```python
# delete redundant columns
sheets = qnetwork_sheets + tab_sheets

for sheet in sheets:
    if "Model Type" in sheet.columns:
        sheet.drop("Model Type", inplace=True, axis=1)

for sheet in sheets:
    if "Unnamed: 0.1" in sheet.columns:
        sheet.drop("Unnamed: 0.1", inplace=True, axis=1)
```

```python
columns = []
for idx, sheet in enumerate(sheets):
    for col in sheet.columns:
        if not col in columns:
            print(f"Adding {col} to columns.")
            columns.append(col)
```

```python
# now we join the spreadsheets together

# first we need to make sure all sheets have the same columns
for sheet in sheets:
    for col in columns:
        if not col in sheet.columns:
            sheet[col] = None
```

```python
master_sheet = pd.concat(sheets, ignore_index=True, sort=True)
```
